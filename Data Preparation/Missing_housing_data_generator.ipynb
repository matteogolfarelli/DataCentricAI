### Loading the Ames Housing Dataset

I am loading the Ames Housing dataset from the given GitHub URL into a pandas DataFrame. Again, I will adjust the URL to its 'raw' version for direct CSV access.

```python
import pandas as pd
ames_csv_url = 'https://raw.githubusercontent.com/matteogolfarelli/DataCentricAI/main/Data%20Preparation/AmesHousing.csv'
df_ames = pd.read_csv(ames_csv_url)
print('Ames Housing Dataset loaded successfully.')

columns_to_retain = ['Order', 'MS Zoning', 'Year Built', 'Overall Qual', 'Overall Cond', 'Lot Area', 'Garage Area', 'Pool Area', 'Bedroom AbvGr', 'Full Bath', 'TotRms AbvGrd', 'SalePrice']
df_ames = df_ames[columns_to_retain]
print('DataFrame filtered to retain specified columns.')
display(df_ames.head())
```

### Statistical Description of the Ames Housing Dataset

This table provides a statistical summary of the numerical columns in the `df_ames` DataFrame.

```python
df_ames.describe(include="all")
df_ames.isna().sum()
```

```python
import numpy as np
import random

# Get the number of rows in the DataFrame
num_rows = len(df_ames)

# Generate 200 unique random indices
random_indices = random.sample(range(num_rows), 197)

# Introduce NaN values at these random indices in 'Year Built' column
df_ames.loc[random_indices, 'Year Built'] = np.nan

print(f"Added 200 missing values to 'Year Built' column.")
# Verify the number of missing values in 'Year Built'
print(f"Number of missing values in 'Year Built' after modification: {df_ames['Year Built'].isnull().sum()}")
```

```python
#import numpy as np
#import random

# Get the number of rows in the DataFrame
num_rows = len(df_ames)

# Generate 270 unique random indices
random_indices_full_bath = random.sample(range(num_rows), 274)

# Introduce NaN values at these random indices in 'Full Bath' column
df_ames.loc[random_indices_full_bath, 'Full Bath'] = np.nan

print(f"Added 270 missing values to 'Full Bath' column.")
# Verify the number of missing values in 'Full Bath'
print(f"Number of missing values in 'Full Bath' after modification: {df_ames['Full Bath'].isnull().sum()}")
```

```python
#import numpy as np
#import random

# Store the initial number of NaNs in 'Overall Qual' to calculate the new ones added
initial_nan_count_overall_qual = df_ames['Overall Qual'].isnull().sum()
print(f"initial number of NaNs in 'Overall Qual' {initial_nan_count_overall_qual}")

# Set to store all unique indices where new NaNs will be introduced
new_nan_indices_overall_qual = set()

# Category 1: Overall Qual < 5 (57 missing values out of 283 - 20%)
eligible_indices_cat1 = df_ames[(df_ames['Overall Qual'] < 5) & (df_ames['Overall Qual'].notna())].index.tolist()
num_to_sample_cat1 = min(57, len(eligible_indices_cat1))
new_nan_indices_overall_qual.update(random.sample(eligible_indices_cat1, num_to_sample_cat1))

# Category 2: 5 <= Overall Qual < 6 (79 missing values out of 792 - 10%)
eligible_indices_cat2 = df_ames[
    (df_ames['Overall Qual'] >= 5) &
    (df_ames['Overall Qual'] < 6) &
    (df_ames['Overall Qual'].notna())
].index.tolist()
num_to_sample_cat2 = min(79, len(eligible_indices_cat2))
new_nan_indices_overall_qual.update(random.sample(eligible_indices_cat2, num_to_sample_cat2))

# Category 3: 6 <= Overall Qual < 8 (66 missing values out of 1314 - 5%)
eligible_indices_cat3 = df_ames[
    (df_ames['Overall Qual'] >= 6) &
    (df_ames['Overall Qual'] < 8) &
    (df_ames['Overall Qual'].notna())
].index.tolist()
num_to_sample_cat3 = min(66, len(eligible_indices_cat3))
new_nan_indices_overall_qual.update(random.sample(eligible_indices_cat3, num_to_sample_cat3))

# Category 4: 8 <= Overall Qual <= 10 (10 missing values out of 478 - 2%)
eligible_indices_cat4 = df_ames[
    (df_ames['Overall Qual'] >= 8) &
    (df_ames['Overall Qual'] <= 10) &
    (df_ames['Overall Qual'].notna())
].index.tolist()
num_to_sample_cat4 = min(10, len(eligible_indices_cat4))
new_nan_indices_overall_qual.update(random.sample(eligible_indices_cat4, num_to_sample_cat4))

# Convert set to list for .loc indexing
indices_to_nan_overall_qual = list(new_nan_indices_overall_qual)

# Introduce NaN values at these selected unique indices
df_ames.loc[indices_to_nan_overall_qual, 'Overall Qual'] = np.nan

# Calculate the number of new NaNs actually added
num_new_nans_added = df_ames['Overall Qual'].isnull().sum() - initial_nan_count_overall_qual

print(f"Attempted to add 120 new missing values to 'Overall Qual' column with specific distribution.")
print(f"Actually added {num_new_nans_added} new missing values based on available eligible data.")
print(f"Total number of missing values in 'Overall Qual' after this modification: {df_ames['Overall Qual'].isnull().sum()}")
```
