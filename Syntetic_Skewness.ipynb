```python
import numpy as np
import pandas as pd
from scipy.stats import skewnorm
import matplotlib.pyplot as plt
import seaborn as sns
```

### Dataset 1: Slight Negative Skewness

To create a dataset with negative skewness, we will use a negative `a` parameter for `skewnorm`. A value of `a` around -5 should provide a slight negative skewness.

```python
# Parameters for Dataset 1 (slight negative skewness)
np.random.seed(42) # for reproducibility
n_points = 200
loc1 = 0 # Mean of the distribution
scale1 = 5 # Standard deviation of the distribution
a1 = -5 # Negative skewness parameter (slight)

# Generate the points for Dataset 1
dataset1 = skewnorm.rvs(a=a1, loc=loc1, scale=scale1, size=n_points)

# Create a DataFrame for better visualization and analysis
df1 = pd.DataFrame({'x': dataset1})

# Calculate mean and standard deviation
mean1 = df1['x'].mean()
std_dev1 = df1['x'].std()

print("### Dataset 1 Statistics (Slight Negative Skewness) ###")
print(df1.describe())
print(f"\nCalculated Skewness: {df1['x'].skew():.4f}")
print(f"Calculated Mean: {mean1:.4f}")
print(f"Calculated Standard Deviation: {std_dev1:.4f}")

# Visualization of Dataset 1
plt.figure(figsize=(10, 6))
sns.histplot(df1['x'], kde=True, bins=15, color='skyblue')
plt.axvline(mean1, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean1:.2f}')
plt.title('Dataset 1 Distribution (Slight Negative Skewness)')
plt.xlabel('x Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Dataset 2: Strong Positive Skewness

To create a dataset with strong positive skewness, we will use a positive `a` parameter with a larger value. A value of `a` around 20-30 should give strong positive skewness.

```python
# Parameters for Dataset 2 (strong positive skewness)
np.random.seed(43) # for reproducibility
loc2 = 0 # Mean of the distribution
scale2 = 5 # Standard deviation of the distribution
a2 = 20 # Positive skewness parameter (strong)

# Generate the points for Dataset 2
dataset2 = skewnorm.rvs(a=a2, loc=loc2, scale=scale2, size=n_points)

# Create a DataFrame for better visualization and analysis
df2 = pd.DataFrame({'x': dataset2})

# Calculate mean and standard deviation
mean2 = df2['x'].mean()
std_dev2 = df2['x'].std()

print("### Dataset 2 Statistics (Strong Positive Skewness) ###")
print(df2.describe())
print(f"\nCalculated Skewness: {df2['x'].skew():.4f}")
print(f"Calculated Mean: {mean2:.4f}")
print(f"Calculated Standard Deviation: {std_dev2:.4f}")

# Visualization of Dataset 2
plt.figure(figsize=(10, 6))
sns.histplot(df2['x'], kde=True, bins=15, color='lightcoral')
plt.axvline(mean2, color='darkblue', linestyle='dashed', linewidth=2, label=f'Mean: {mean2:.2f}')
plt.title('Dataset 2 Distribution (Strong Positive Skewness)')
plt.xlabel('x Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
