```python
from google.colab import files
import pandas as pd
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np

#------------------------------------------------------------- Data loading and Data frame creation
url = "https://raw.githubusercontent.com/matteogolfarelli/DataCentricAI/main/CarEvaluation.csv"
df = pd.read_csv(url)

print("The dataset contains ", len(df), "tuple")
print("\n")
print(df.head(), df.tail()) # Show the first five lines
print("\n")

#------------------------------------------------------------- Data preprocessing
target_col = "class"

X = df.drop(columns=target_col)
y = df[target_col]
print("\n ")
print(df.head)
print("\n ------------------- Features: sample tuples \n")
print(X.head)
print("\n ------------------- Class: sample tuples \n")
print(y.head)
```

```python
# Apply Label Encoding to each individual column of X

from sklearn.preprocessing import LabelEncoder
import pandas as pd

X_encoded = pd.DataFrame()
for column in X.columns:
    encoder = LabelEncoder()
    X_encoded[column] = encoder.fit_transform(X[column])
```

```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd
import numpy as np

# Define the ordered categories for each feature
categories = [
    ['low', 'med', 'high', 'vhigh'],  # buying
    ['low', 'med', 'high', 'vhigh'],  # maint
    ['2', '3', '4', '5more'],      # doors
    ['2', '4', 'more'],             # persons
    ['small', 'med', 'big'],        # lug_boot
    ['low', 'med', 'high']          # safety
]

# Initialize OrdinalEncoder with the specified categories
# handle_unknown='use_encoded_value' and unknown_value=np.nan allow for handling values not in categories
encoder = OrdinalEncoder(categories=categories, handle_unknown='use_encoded_value', unknown_value=np.nan)

# Apply Ordinal Encoding to the X DataFrame
X_encoded = pd.DataFrame(encoder.fit_transform(X), columns=X.columns)
```

```python
print("\n ------------------- Encoded Features (Ordinal): sample tuples \n")
print(X_encoded.head())
```

```python
# Apply OneHot Encoding to categorical features in X

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
X_encoded = encoder.fit_transform(X)
```

```python
print("\n ------------------- Encoded Features: sample tuples \n")
print(X_encoded)
```

```python
#------------------------------------------------------------- Decision tree: Model training and testing with split

from sklearn.tree import DecisionTreeClassifier, plot_tree

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=42, stratify=y)

clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# Predizioni sul test set
y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred, average='macro'))
print("Recall: ", recall_score(y_test, y_pred, average='macro'))
print("F1-score: ", f1_score(y_test, y_pred, average='macro'))
```

```python
from sklearn.neighbors import KNeighborsClassifier

#------------------------------------------------------------- KNearestNeighbors: Model training and testing with split
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=42, stratify=y)

clf = KNeighborsClassifier()
clf.fit(X_train, y_train)

# Predizioni sul test set
y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred, average='macro'))
print("Recall: ", recall_score(y_test, y_pred, average='macro'))
print("F1-score: ", f1_score(y_test, y_pred, average='macro'))
```
