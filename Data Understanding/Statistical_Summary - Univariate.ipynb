### Load Bank Data dataset

First, I will install `scipy` and `pandas` if they are not already installed, then download the `.arff` file from the provided GitHub link. After that, I'll load the dataset into a pandas DataFrame and display its head.

```python
import requests
import pandas as pd
from scipy.io import arff
import io

# Raw GitHub URL for the dataset
github_url = "https://github.com/matteogolfarelli/DataCentricAI/raw/main/Data%20Understanding/bank-data.arff"

# Download the file content
response = requests.get(github_url)
response.raise_for_status() # Raise an exception for HTTP errors

# Load the ARFF data
data, meta = arff.loadarff(io.StringIO(response.text))

# Convert to pandas DataFrame
df = pd.DataFrame(data)

for column in df.select_dtypes(include=['object']).columns:
    df[column] = df[column].apply(lambda x: x.decode('utf-8'))

# Display the first 5 rows of the DataFrame
display(df.head(),df.tail())
```

The `b'` prefix in the DataFrame output signifies that the values are byte strings. This is a common occurrence when loading data from formats like ARFF. To address this and ensure better readability, the next step involves converting these byte strings into standard Python strings.

To achieve this, I'll iterate through all columns with the `object` data type, which typically represent categorical or text data. For each column, I'll apply a lambda function to decode the byte strings using UTF-8 encoding. This process will transform the byte-prefixed values into clean, human-readable strings, making the DataFrame easier to work with and understand for subsequent analysis.

### Categorical Feature Analysis

Now that the byte strings have been decoded, I will proceed with a detailed analysis of the categorical features. For each categorical attribute, I will compute and display the following:

1.  **Frequency Count**: The number of occurrences for each unique value.
2.  **Bar Chart**: A visual representation of the frequency counts.
3.  **Relative Frequency**: The proportion of each unique value relative to the total number of entries.
4.  **Mode**: The most frequently occurring value(s).
5.  **Number of Unique Values**: The count of distinct categories.
6.  **Number of Missing Values**: The count of `NaN` or null entries.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify categorical columns (object type after decoding)
categorical_cols = df.select_dtypes(include=['object']).columns

print(f"Categorical columns identified: {list(categorical_cols)}")

for col in categorical_cols:
    print(f"\n--- Analysis for column: '{col}' ---")

    # 1. Frequency Count
    freq_count = df[col].value_counts()
    print("Frequency Count:")
    display(freq_count)

    # 2. Bar Chart
    plt.figure(figsize=(8, 5))
    sns.barplot(x=freq_count.index, y=freq_count.values, hue=freq_count.index, palette='viridis', legend=False)
    plt.title(f'Frequency Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # 3. Relative Frequency
    relative_freq = df[col].value_counts(normalize=True)
    print("Relative Frequency:")
    display(relative_freq)

    # 4. Mode
    mode_value = df[col].mode()
    print("Mode:")
    display(mode_value)

    # 5. Number of Unique Values
    num_unique = df[col].nunique()
    print(f"Number of Unique Values: {num_unique}")

    # 6. Number of Missing Values
    missing_values = df[col].isnull().sum()
    print(f"Number of Missing Values: {missing_values}")
```

### Numerical Feature Analysis

Next, I will analyze the numerical features in the dataset. For each numerical attribute, I will compute and display the following:

1.  **Mean**: The average value.
2.  **Median**: The middle value when sorted.
3.  **Mode**: The most frequently occurring value(s).
4.  **Standard Deviation**: A measure of the dispersion of a dataset relative to its mean.
5.  **Variance**: The average of the squared differences from the mean.
6.  **Range**: The difference between the maximum and minimum values.
7.  **Interquartile Range (IQR)**: The range between the first and third quartiles.
8.  **Skewness**: A measure of the asymmetry of the probability distribution.
9.  **Kurtosis**: A measure of the 'tailedness' of the probability distribution.
10. **Number of Missing Values**: The count of `NaN` or null entries.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numerical columns
numerical_cols = df.select_dtypes(include=np.number).columns

print(f"Numerical columns identified: {list(numerical_cols)}")

for col in numerical_cols:
    print(f"\n--- Analysis for column: '{col}' ---")

    # 1. Mean
    mean_value = df[col].mean()
    print(f"Mean: {mean_value:.2f}")

    # 2. Median
    median_value = df[col].median()
    print(f"Median: {median_value:.2f}")

    # 3. Mode
    mode_value = df[col].mode()
    print("Mode:")
    display(mode_value)

    # 4. Standard Deviation
    std_dev_value = df[col].std()
    print(f"Standard Deviation: {std_dev_value:.2f}")

    # 5. Variance
    variance_value = df[col].var()
    print(f"Variance: {variance_value:.2f}")

    # 6. Range
    range_value = df[col].max() - df[col].min()
    print(f"Range: {range_value:.2f}")

    # 7. Interquartile Range (IQR)
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    iqr_value = Q3 - Q1
    print(f"IQR: {iqr_value:.2f}")

    # 8. Skewness
    skew_value = df[col].skew()
    print(f"Skewness: {skew_value:.2f}")

    # 9. Kurtosis
    kurt_value = df[col].kurtosis()
    print(f"Kurtosis: {kurt_value:.2f}")

    # 10. Number of Missing Values
    missing_values = df[col].isnull().sum()
    print(f"Number of Missing Values: {missing_values}")

    # Plot data distribution (Histogram)
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

### Percentile Analysis and Box Plots for Numerical Features

This section focuses on visualizing the distribution and percentiles of numerical features using box plots, and explicitly computing the quartiles (Q1, Median, Q3).

1.  **Quartiles (Q1, Q2, Q3)**: Dividing the data into four equal parts.
2.  **Box Plot**: A standardized way of displaying the dataset based on a five-number summary: the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numerical columns
numerical_cols = df.select_dtypes(include=np.number).columns

print(f"Numerical columns identified for percentile analysis: {list(numerical_cols)}")

for col in numerical_cols:
    print(f"\n--- Percentile Analysis for column: '{col}' ---")

    # Calculate Quartiles
    Q1 = df[col].quantile(0.25)
    Q2 = df[col].median() # Median is Q2
    Q3 = df[col].quantile(0.75)

    print(f"Q1 (25th Percentile): {Q1:.2f}")
    print(f"Median (Q2 - 50th Percentile): {Q2:.2f}")
    print(f"Q3 (75th Percentile): {Q3:.2f}")

    # Generate Box Plot
    plt.figure(figsize=(8, 5))
    sns.boxplot(y=df[col], color='lightcoral')
    plt.title(f'Box Plot of {col} (Percentiles)')
    plt.ylabel(col)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```
